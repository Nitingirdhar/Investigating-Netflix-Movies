```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Start coding!
netflix_df = pd.read_csv("netflix_data.csv")
```

```python
netflix_df.head()
```

```python
# Subset the DataFram for the type "Movie"
netflix_subset = netflix_df[netflix_df["type"] == "Movie"]
```

```python
#Select only the columns of intreset 
netflix_movies = netflix_subset[["title", "country", "genre", "release_year", "duration"]]
```

```python
# Filter for duration shorter than 60 min
short_movies = netflix_movies[netflix_movies.duration < 60]

#Shortest Movies
short_movies
```

```python
# An empty colors list 
colors = [] 
```

```python
# Iterate over rows of netflix_movies
for label, row in netflix_movies.iterrows() :
    if row["genre"] == "Children" :
        colors.append("red")
    elif row["genre"] == "Documentaries" :
        colors.append("blue")
    elif row["genre"] == "Stand-Up":
        colors.append("green")
    else:
        colors.append("black")
```

```python
#Inspect the first 10 values in your list 
colors[:10]
```

```python
#Creating a scatter plot chart
#Set the figure style and initalize a new figure
fig = plt.figure(figsize=(12, 8))

# Scatter plot of duration vs release year
plt.scatter(netflix_movies.release_year, netflix_movies.duration, c=colors[:len(netflix_movies)])  # Adjusted the 'c' argument to match the size of 'x' and 'y'
# Create a scatter plot of duration versus release_year
plt.scatter(netflix_movies.release_year, netflix_movies.duration, c=colors)

# Create a title and axis labels
plt.title("Movie Duration by Year of Release")
plt.xlabel("Release year")
plt.ylabel("Duration (min)")

# Show the plot
plt.show()
```
